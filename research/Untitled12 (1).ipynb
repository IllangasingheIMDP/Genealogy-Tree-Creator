```python
#!/usr/bin/env python3
"""
Improved Genealogy Relationship Validator
Enhanced BERT model for validating family relationships with better training strategies
"""

# =============================================================================
# INSTALLATION AND IMPORTS
# =============================================================================

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    TrainingArguments, Trainer, EarlyStoppingCallback
)
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Tuple, Optional
import json
import warnings
import time
from datetime import datetime
warnings.filterwarnings('ignore')

# Check device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# =============================================================================
# ENHANCED DATASET PROCESSING
# =============================================================================

class ImprovedGenealogyDataset(Dataset):
    """Enhanced dataset with better tokenization and data augmentation"""

    def __init__(self, texts: List[str], labels: List[int], tokenizer, max_length: int = 512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]

        # Enhanced tokenization with proper handling
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt',
            add_special_tokens=True
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# =============================================================================
# DATA AUGMENTATION AND EXPANSION
# =============================================================================

def augment_dataset(df: pd.DataFrame) -> pd.DataFrame:
    """
    Significantly expand the dataset with variations and synthetic examples
    """
    print("Starting data augmentation...")
    augmented_data = []

    # Add original data
    for _, row in df.iterrows():
        augmented_data.append({
            'input_text': row['input_text'],
            'label': 1 if row['is_correct'] else 0,
            'person1': row['person1'],
            'relation': row['relation'],
            'person2': row['person2']
        })

    # Create relationship variations for positive examples
    positive_df = df[df['is_correct'] == True]

    for _, row in positive_df.iterrows():
        context = row['wikipedia_context']
        person1, relation, person2 = row['person1'], row['relation'], row['person2']

        # Alternative question formats
        variations = [
            f"Given the context: {context} Is it true that {person1} is {relation} {person2}?",
            f"Context: {context} Verify: {person1} {relation} {person2}",
            f"Based on this information: {context} Question: {person1} {relation} {person2} - True or False?",
            f"Text: {context} Relationship to verify: Does {person1} have the relationship '{relation}' with {person2}?"
        ]

        for variation in variations:
            augmented_data.append({
                'input_text': variation,
                'label': 1,
                'person1': person1,
                'relation': relation,
                'person2': person2
            })

        # Create reverse relationships where applicable
        if relation == "child of":
            reverse_context = context.replace(f"{person1}", "[PERSON1]").replace(f"{person2}", "[PERSON2]")
            reverse_context = reverse_context.replace("[PERSON2]", f"{person2}").replace("[PERSON1]", f"{person1}")

            reverse_text = f"Context: {reverse_context} Question: Is '{person2} parent of {person1}' correct?"
            augmented_data.append({
                'input_text': reverse_text,
                'label': 1,
                'person1': person2,
                'relation': "parent of",
                'person2': person1
            })

    # Add more negative examples by mixing people incorrectly
    all_people = list(set(df['person1'].tolist() + df['person2'].tolist()))
    relations = ["child of", "spouse of", "parent of"]

    # Generate 100 negative examples
    import random
    random.seed(42)
    negative_count = 0
    attempts = 0
    max_attempts = 500

    while negative_count < 100 and attempts < max_attempts:
        attempts += 1
        person1 = random.choice(all_people)
        person2 = random.choice(all_people)
        relation = random.choice(relations)

        # Skip if this is actually a correct relationship in our dataset
        is_real_relationship = any(
            (row['person1'] == person1 and row['person2'] == person2 and row['relation'] == relation and row['is_correct'])
            for _, row in df.iterrows()
        )

        if not is_real_relationship and person1 != person2:
            negative_context = f"There is no evidence or documentation that {person1} has any {relation} relationship with {person2}."
            negative_text = f"Context: {negative_context} Question: Is '{person1} {relation} {person2}' correct?"

            augmented_data.append({
                'input_text': negative_text,
                'label': 0,
                'person1': person1,
                'relation': relation,
                'person2': person2
            })
            negative_count += 1

    print(f"Original dataset size: {len(df)}")
    print(f"Augmented dataset size: {len(augmented_data)}")

    return pd.DataFrame(augmented_data)

# =============================================================================
# ENHANCED TRAINING WITH BETTER PARAMETERS
# =============================================================================

def compute_metrics(eval_pred):
    """Compute additional metrics during training"""
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)

    accuracy = accuracy_score(labels, predictions)

    # Calculate probabilities for AUC
    try:
        probs = torch.softmax(torch.tensor(eval_pred[0]), dim=1)[:, 1].numpy()
        auc = roc_auc_score(labels, probs)
    except:
        auc = 0.0

    return {
        'accuracy': accuracy,
        'auc': auc
    }

def train_improved_model(df: pd.DataFrame,
                        model_name: str = "distilbert-base-uncased",
                        num_epochs: int = 10,
                        learning_rate: float = 2e-5,
                        batch_size: int = 16,
                        max_length: int = 512) -> Tuple:
    """
    Train with improved parameters and validation
    """
    print(f"Training improved model with {len(df)} examples...")

    # Prepare data
    texts = df['input_text'].tolist()
    labels = df['label'].tolist()

    # Stratified split to maintain label balance
    train_texts, val_texts, train_labels, val_labels = train_test_split(
        texts, labels,
        test_size=0.2,
        random_state=42,
        stratify=labels
    )

    print(f"Training samples: {len(train_texts)} (Positive: {sum(train_labels)}, Negative: {len(train_texts) - sum(train_labels)})")
    print(f"Validation samples: {len(val_texts)} (Positive: {sum(val_labels)}, Negative: {len(val_texts) - sum(val_labels)})")

    # Initialize model with better configuration
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=2,
        hidden_dropout_prob=0.1,
        attention_probs_dropout_prob=0.1
    ).to(device)

    # Create datasets
    train_dataset = ImprovedGenealogyDataset(train_texts, train_labels, tokenizer, max_length)
    val_dataset = ImprovedGenealogyDataset(val_texts, val_labels, tokenizer, max_length)

    # Enhanced training arguments
    training_args = TrainingArguments(
        output_dir='./genealogy-validator-improved',
        num_train_epochs=num_epochs,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        weight_decay=0.01,
        warmup_steps=100,

        # Evaluation and logging
        eval_strategy="steps",
        eval_steps=50,
        save_steps=100,
        logging_steps=25,

        # Early stopping and model selection
        load_best_model_at_end=True,
        metric_for_best_model="eval_accuracy",
        greater_is_better=True,

        # Learning rate scheduling
        lr_scheduler_type="cosine",

        # Mixed precision for faster training
        fp16=torch.cuda.is_available(),

        # Gradient accumulation for effective larger batch size
        gradient_accumulation_steps=2,

        # Prevent overfitting
        save_total_limit=3,

        # Reproducibility
        seed=42,
        data_seed=42,
    )

    # Initialize trainer with early stopping
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        tokenizer=tokenizer,
        compute_metrics=compute_metrics,
        callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
    )

    # Record training start time
    start_time = time.time()
    print(f"Training started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

    # Train the model
    trainer.train()

    # Calculate training time
    training_time = time.time() - start_time
    print(f"Training completed in: {training_time:.2f} seconds ({training_time/60:.2f} minutes)")

    # Save the final model
    model.save_pretrained('./genealogy-validator-improved-final')
    tokenizer.save_pretrained('./genealogy-validator-improved-final')

    # Evaluate on validation set
    eval_results = trainer.evaluate()
    print("\nValidation Results:")
    for key, value in eval_results.items():
        print(f"  {key}: {value:.4f}")

    return model, tokenizer, val_texts, val_labels, eval_results

# =============================================================================
# IMPROVED VALIDATION PIPELINE
# =============================================================================

class ImprovedGenealogyValidator:
    """Enhanced validation pipeline with better confidence handling"""

    def __init__(self, model_path: str = "./genealogy-validator-improved-final"):
        print(f"Loading model from {model_path}...")

        try:
            self.tokenizer = AutoTokenizer.from_pretrained(model_path)
            self.model = AutoModelForSequenceClassification.from_pretrained(model_path).to(device)
            self.model.eval()
            print("Model loaded successfully!")
        except Exception as e:
            print(f"Error loading model: {e}")
            print("Make sure to train the model first!")
            raise

    def validate_single_relationship(self,
                                   relationship_text: str,
                                   context: str,
                                   confidence_threshold: float = 0.6) -> Dict:
        """
        Validate a single relationship with improved confidence handling
        """
        # Format input text
        input_text = f"Context: {context} Question: Is '{relationship_text}' correct?"

        # Tokenize
        inputs = self.tokenizer(
            input_text,
            truncation=True,
            padding='max_length',
            max_length=512,
            return_tensors='pt'
        ).to(device)

        # Get prediction
        with torch.no_grad():
            outputs = self.model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=-1)

            # Get prediction and confidence
            predicted_class = probabilities.argmax().item()
            confidence = probabilities.max().item()

            # Get probability for positive class specifically
            positive_prob = probabilities[0][1].item()

        # Determine status
        if confidence >= confidence_threshold:
            status = "VALID" if predicted_class == 1 else "INVALID"
        else:
            status = "UNCERTAIN"

        return {
            'relationship': relationship_text,
            'is_valid': predicted_class == 1,
            'confidence': confidence,
            'positive_probability': positive_prob,
            'status': status,
            'input_text': input_text
        }

    def validate_batch_relationships(self,
                                   relationships: List[str],
                                   contexts: List[str],
                                   confidence_threshold: float = 0.6) -> Dict:
        """
        Validate multiple relationships efficiently
        """
        results = []

        print(f"Validating {len(relationships)} relationships...")

        for i, (rel, context) in enumerate(zip(relationships, contexts)):
            result = self.validate_single_relationship(rel, context, confidence_threshold)
            results.append(result)

            if (i + 1) % 10 == 0:
                print(f"Processed {i + 1}/{len(relationships)} relationships")

        # Calculate summary statistics
        total = len(results)
        valid_count = sum(1 for r in results if r['status'] == 'VALID')
        invalid_count = sum(1 for r in results if r['status'] == 'INVALID')
        uncertain_count = sum(1 for r in results if r['status'] == 'UNCERTAIN')

        avg_confidence = np.mean([r['confidence'] for r in results])
        avg_positive_prob = np.mean([r['positive_probability'] for r in results])

        return {
            'results': results,
            'summary': {
                'total_relationships': total,
                'valid': valid_count,
                'invalid': invalid_count,
                'uncertain': uncertain_count,
                'avg_confidence': avg_confidence,
                'avg_positive_probability': avg_positive_prob,
                'accuracy_if_all_correct': valid_count / total * 100 if total > 0 else 0
            }
        }

# =============================================================================
# EVALUATION AND TESTING
# =============================================================================

def evaluate_model_performance(model_path: str, test_df: pd.DataFrame) -> Dict:
    """
    Comprehensive model evaluation
    """
    print("\nEvaluating model performance...")

    validator = ImprovedGenealogyValidator(model_path)

    # Prepare test data
    relationships = []
    contexts = []
    true_labels = []

    for _, row in test_df.iterrows():
        rel_text = f"{row['person1']} {row['relation']} {row['person2']}"
        relationships.append(rel_text)
        contexts.append(row['wikipedia_context'])
        true_labels.append(1 if row['is_correct'] else 0)

    # Get predictions
    results = validator.validate_batch_relationships(relationships, contexts, confidence_threshold=0.6)

    # Calculate metrics
    predicted_labels = [1 if r['is_valid'] else 0 for r in results['results']]
    confidences = [r['confidence'] for r in results['results']]

    accuracy = accuracy_score(true_labels, predicted_labels)

    # Classification report
    report = classification_report(true_labels, predicted_labels, output_dict=True)

    print(f"\nTest Accuracy: {accuracy:.4f}")
    print(f"Average Confidence: {np.mean(confidences):.4f}")
    print(f"Precision (Class 1): {report['1']['precision']:.4f}")
    print(f"Recall (Class 1): {report['1']['recall']:.4f}")
    print(f"F1-Score (Class 1): {report['1']['f1-score']:.4f}")

    return {
        'accuracy': accuracy,
        'classification_report': report,
        'predictions': results
    }

# =============================================================================
# MAIN EXECUTION
# =============================================================================

def main():
    print("IMPROVED GENEALOGY RELATIONSHIP VALIDATOR")
    print("=" * 80)

    # Step 1: Load your CSV data
    print("Step 1: Loading dataset from CSV...")

    # You can load your CSV file here
    # df = pd.read_csv('your_dataset.csv')

    # For demonstration, I'll create a sample based on your data structure
    sample_data = {
        'person1': ['Albert Einstein', 'Albert Einstein', 'Charles III', 'Barack Obama', 'Albert Einstein'],
        'relation': ['child of', 'spouse of', 'child of', 'spouse of', 'child of'],
        'person2': ['Hermann Einstein', 'Mileva Marić', 'Queen Elizabeth II', 'Michelle Obama', 'Isaac Newton'],
        'is_correct': [True, True, True, True, False],
        'wikipedia_context': [
            'Albert Einstein was born in Ulm, in the Kingdom of Württemberg in the German Empire, on 14 March 1879. His parents, secular Ashkenazi Jews, were Hermann Einstein, a salesman and engineer, and Pauline Koch.',
            'Einstein and Marić married in January 1903. In May 1904, their son Hans Albert was born in Bern, Switzerland.',
            'Charles was born during the reign of his maternal grandfather, King George VI, as the first child of Princess Elizabeth, Duchess of Edinburgh (later Queen Elizabeth II), and Philip, Duke of Edinburgh.',
            'In June 1989, Obama met Michelle Robinson when he was employed at Sidley Austin. They began dating later that summer, became engaged in 1991, and were married on October 3, 1992.',
            'Albert Einstein was born in Ulm, in the Kingdom of Württemberg in the German Empire, on 14 March 1879. Isaac Newton lived from 1643-1727, nearly 150 years before Einstein was born.'
        ]
    }

    df = pd.DataFrame(sample_data)

    print(f"Loaded dataset with {len(df)} examples")
    print(f"Positive examples: {sum(df['is_correct'])}")
    print(f"Negative examples: {len(df) - sum(df['is_correct'])}")

    # Step 2: Augment the dataset
    print("\nStep 2: Augmenting dataset...")

    # Add input_text column if not present
    if 'input_text' not in df.columns:
        df['input_text'] = df.apply(
            lambda row: f"Context: {row['wikipedia_context']} Question: Is '{row['person1']} {row['relation']} {row['person2']}' correct?",
            axis=1
        )

    augmented_df = augment_dataset(df)

    # Step 3: Train the improved model
    print("\nStep 3: Training improved model...")
    print("This will take longer but should produce much better results...")

    try:
        model, tokenizer, val_texts, val_labels, eval_results = train_improved_model(
            augmented_df,
            num_epochs=10,  # Increased from 3
            learning_rate=2e-5,  # Optimized learning rate
            batch_size=16,
            max_length=512  # Increased context length
        )

        print("\nTraining completed successfully!")

        # Step 4: Test the model
        print("\nStep 4: Testing improved model...")

        validator = ImprovedGenealogyValidator()

        # Test on original examples
        test_relationships = [
            "Albert Einstein child of Hermann Einstein",
            "Albert Einstein spouse of Mileva Marić",
            "Charles III child of Queen Elizabeth II",
            "Barack Obama spouse of Michelle Obama",
            "Albert Einstein child of Isaac Newton"  # Should be invalid
        ]

        test_contexts = [
            "Albert Einstein was born in Ulm, in the Kingdom of Württemberg in the German Empire, on 14 March 1879. His parents, secular Ashkenazi Jews, were Hermann Einstein, a salesman and engineer, and Pauline Koch.",
            "Einstein and Marić married in January 1903. In May 1904, their son Hans Albert was born in Bern, Switzerland.",
            "Charles was born during the reign of his maternal grandfather, King George VI, as the first child of Princess Elizabeth, Duchess of Edinburgh (later Queen Elizabeth II), and Philip, Duke of Edinburgh.",
            "In June 1989, Obama met Michelle Robinson when he was employed at Sidley Austin. They began dating later that summer, became engaged in 1991, and were married on October 3, 1992.",
            "Albert Einstein was born in Ulm, in the Kingdom of Württemberg in the German Empire, on 14 March 1879. Isaac Newton lived from 1643-1727, nearly 150 years before Einstein was born."
        ]

        results = validator.validate_batch_relationships(test_relationships, test_contexts)

        print("\n" + "=" * 80)
        print("IMPROVED VALIDATION RESULTS")
        print("=" * 80)

        for result in results['results']:
            print(f"\nRelationship: {result['relationship']}")
            print(f"Status: {result['status']}")
            print(f"Confidence: {result['confidence']:.4f}")
            print(f"Positive Probability: {result['positive_probability']:.4f}")

        print(f"\nSummary Statistics:")
        summary = results['summary']
        for key, value in summary.items():
            if isinstance(value, float):
                print(f"  {key}: {value:.4f}")
            else:
                print(f"  {key}: {value}")

    except Exception as e:
        print(f"Error during training: {e}")
        print("This might be due to insufficient data or system limitations.")

if __name__ == "__main__":
    main()

# =============================================================================
# TIME ESTIMATION FUNCTION
# =============================================================================

def estimate_training_time():
    """
    Estimate training time based on system capabilities
    """
    print("\nTRAINING TIME ESTIMATION")
    print("=" * 50)

    if torch.cuda.is_available():
        gpu_name = torch.cuda.get_device_name(0)
        print(f"GPU Available: {gpu_name}")
        print("Estimated training time with GPU:")
        print("  - With your 33 examples: 2-5 minutes")
        print("  - With augmented dataset (~500 examples): 10-15 minutes")
        print("  - With 10 epochs: 15-25 minutes total")
    else:
        print("CPU Only Training")
        print("Estimated training time with CPU:")
        print("  - With your 33 examples: 10-15 minutes")
        print("  - With augmented dataset (~500 examples): 30-45 minutes")
        print("  - With 10 epochs: 45-75 minutes total")

    print("\nFactors affecting training time:")
    print("  - Dataset size (more examples = longer training)")
    print("  - Number of epochs (more epochs = longer training)")
    print("  - Batch size (larger batches = faster per epoch)")
    print("  - Model size (DistilBERT is relatively fast)")
    print("  - Hardware (GPU vs CPU makes huge difference)")

# Run time estimation
if __name__ == "__main__":
    estimate_training_time()
```

```python
#!/usr/bin/env python3
"""
Similarity-Based Genealogy Validator
Uses semantic similarity + lightweight classification instead of heavy fine-tuning
"""

import requests
from bs4 import BeautifulSoup
import re
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd
from typing import List, Dict, Tuple, Optional
import warnings
warnings.filterwarnings('ignore')

class WikipediaRetriever:
    """Fetch and process Wikipedia content"""

    def __init__(self):
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        })

    def get_wikipedia_text(self, person_name: str) -> str:
        """Get Wikipedia article text for a person"""
        try:
            # Try to get Wikipedia page
            search_url = f"https://en.wikipedia.org/api/rest_v1/page/summary/{person_name.replace(' ', '_')}"

            response = self.session.get(search_url, timeout=10)
            if response.status_code == 200:
                data = response.json()
                extract = data.get('extract', '')

                # Get full article content
                page_url = f"https://en.wikipedia.org/wiki/{person_name.replace(' ', '_')}"
                page_response = self.session.get(page_url, timeout=10)

                if page_response.status_code == 200:
                    soup = BeautifulSoup(page_response.content, 'html.parser')

                    # Extract main content paragraphs
                    content_div = soup.find('div', {'id': 'mw-content-text'})
                    if content_div:
                        paragraphs = content_div.find_all('p')
                        full_text = ' '.join([p.get_text().strip() for p in paragraphs[:10]])  # First 10 paragraphs
                        return full_text

                return extract
            else:
                print(f"Could not fetch Wikipedia page for {person_name}")
                return ""

        except Exception as e:
            print(f"Error fetching Wikipedia for {person_name}: {e}")
            return ""

    def chunk_text(self, text: str, chunk_size: int = 300, overlap: int = 50) -> List[str]:
        """Split text into overlapping chunks"""
        if not text:
            return []

        words = text.split()
        chunks = []

        for i in range(0, len(words), chunk_size - overlap):
            chunk = ' '.join(words[i:i + chunk_size])
            if len(chunk.strip()) > 50:  # Only keep substantial chunks
                chunks.append(chunk.strip())

        return chunks

class SimilarityBasedValidator:
    """Validate relationships using semantic similarity + lightweight classification"""

    def __init__(self, use_pretrained_classifier: bool = False):
        print("Initializing Similarity-Based Validator...")

        # Load sentence transformer for semantic similarity
        self.similarity_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
        print("✓ Loaded sentence transformer model")

        # Load Wikipedia retriever
        self.wiki_retriever = WikipediaRetriever()
        print("✓ Initialized Wikipedia retriever")

        # Optionally load your trained classifier
        self.classifier = None
        self.tokenizer = None

        if use_pretrained_classifier:
            try:
                self.tokenizer = AutoTokenizer.from_pretrained('./genealogy-validator-fixed-final')
                self.classifier = AutoModelForSequenceClassification.from_pretrained('./genealogy-validator-fixed-final')
                self.classifier.eval()
                print("✓ Loaded pre-trained classifier")
            except:
                print("! Could not load pre-trained classifier, using similarity-only approach")

        print("Initialization complete!")

    def find_relevant_chunks(self, relationship_text: str, article_chunks: List[str], top_k: int = 3) -> List[Tuple[str, float]]:
        """Find most relevant chunks using semantic similarity"""
        if not article_chunks:
            return []

        # Create embeddings
        relationship_embedding = self.similarity_model.encode([relationship_text])
        chunk_embeddings = self.similarity_model.encode(article_chunks)

        # Calculate similarities
        similarities = cosine_similarity(relationship_embedding, chunk_embeddings)[0]

        # Get top k most similar chunks
        top_indices = np.argsort(similarities)[-top_k:][::-1]

        relevant_chunks = []
        for idx in top_indices:
            if similarities[idx] > 0.1:  # Only include chunks with reasonable similarity
                relevant_chunks.append((article_chunks[idx], similarities[idx]))

        return relevant_chunks

    def extract_relationship_keywords(self, person1: str, relation: str, person2: str) -> List[str]:
        """Extract keywords that indicate the relationship"""

        relation_keywords = {
            'child of': [
                f"{person1} was born", f"son of {person2}", f"daughter of {person2}",
                f"{person1} born", f"parents {person2}", f"{person2} and",
                f"father {person2}", f"mother {person2}", f"{person2} had", "born to"
            ],
            'spouse of': [
                f"{person1} married {person2}", f"married {person2}", f"wedding",
                f"husband {person2}", f"wife {person2}", f"{person1} and {person2} married",
                f"divorce", f"spouse", f"marriage"
            ],
            'parent of': [
                f"{person2} was born", f"son {person2}", f"daughter {person2}",
                f"{person1} had", f"children", f"father of {person2}", f"mother of {person2}"
            ]
        }

        return relation_keywords.get(relation, [relation])

    def similarity_score(self, relationship_text: str, context_chunks: List[str]) -> float:
        """Calculate overall similarity score between relationship and context"""
        if not context_chunks:
            return 0.0

        # Get the best chunk
        relevant_chunks = self.find_relevant_chunks(relationship_text, context_chunks, top_k=1)

        if relevant_chunks:
            return relevant_chunks[0][1]  # Return the highest similarity score

        return 0.0

    def keyword_match_score(self, person1: str, relation: str, person2: str, context_chunks: List[str]) -> float:
        """Calculate keyword-based matching score"""
        if not context_chunks:
            return 0.0

        keywords = self.extract_relationship_keywords(person1, relation, person2)
        context_text = ' '.join(context_chunks).lower()

        matches = 0
        for keyword in keywords:
            if keyword.lower() in context_text:
                matches += 1

        return matches / len(keywords) if keywords else 0.0

    def validate_relationship_with_article(self,
                                         person1: str,
                                         relation: str,
                                         person2: str,
                                         article_text: str = None,
                                         confidence_threshold: float = 0.3) -> Dict:
        """
        Validate a relationship using similarity-based approach
        """

        # If no article text provided, try to fetch from Wikipedia
        if not article_text:
            print(f"Fetching Wikipedia content for {person1}...")
            article_text = self.wiki_retriever.get_wikipedia_text(person1)

            if not article_text:
                print(f"Trying Wikipedia for {person2}...")
                article_text = self.wiki_retriever.get_wikipedia_text(person2)

        if not article_text:
            return {
                'relationship': f"{person1} {relation} {person2}",
                'is_valid': False,
                'confidence': 0.0,
                'method': 'no_context',
                'status': 'NO_CONTEXT',
                'evidence': []
            }

        # Split article into chunks
        chunks = self.wiki_retriever.chunk_text(article_text)

        # Create relationship query
        relationship_text = f"{person1} {relation} {person2}"

        # Find relevant chunks
        relevant_chunks = self.find_relevant_chunks(relationship_text, chunks, top_k=3)

        # Calculate scores
        similarity_score = self.similarity_score(relationship_text, chunks)
        keyword_score = self.keyword_match_score(person1, relation, person2, chunks)

        # Combined score (weighted average)
        combined_score = (similarity_score * 0.7) + (keyword_score * 0.3)

        # Determine validity
        is_valid = combined_score >= confidence_threshold

        # Determine status
        if combined_score >= 0.5:
            status = "VALID"
        elif combined_score >= confidence_threshold:
            status = "LIKELY_VALID"
        elif combined_score >= 0.1:
            status = "UNCERTAIN"
        else:
            status = "INVALID"

        return {
            'relationship': relationship_text,
            'is_valid': is_valid,
            'confidence': combined_score,
            'similarity_score': similarity_score,
            'keyword_score': keyword_score,
            'method': 'similarity_based',
            'status': status,
            'evidence': [chunk for chunk, score in relevant_chunks]
        }

    def validate_multiple_relationships(self,
                                      relationships: List[Tuple[str, str, str]],
                                      person_articles: Dict[str, str] = None,
                                      confidence_threshold: float = 0.3) -> Dict:
        """
        Validate multiple relationships efficiently
        """
        results = []

        # Cache for Wikipedia articles
        article_cache = person_articles or {}

        print(f"Validating {len(relationships)} relationships...")

        for i, (person1, relation, person2) in enumerate(relationships):
            print(f"Processing {i+1}/{len(relationships)}: {person1} {relation} {person2}")

            # Try to use cached articles first
            article_text = None
            if person1 in article_cache:
                article_text = article_cache[person1]
            elif person2 in article_cache:
                article_text = article_cache[person2]

            result = self.validate_relationship_with_article(
                person1, relation, person2, article_text, confidence_threshold
            )
            results.append(result)

            # Cache the article if we fetched it
            if not article_text and result['evidence']:
                article_cache[person1] = ' '.join(result['evidence'])

        # Calculate summary statistics
        total = len(results)
        valid_count = sum(1 for r in results if r['status'] in ['VALID', 'LIKELY_VALID'])
        invalid_count = sum(1 for r in results if r['status'] == 'INVALID')
        uncertain_count = sum(1 for r in results if r['status'] == 'UNCERTAIN')
        no_context_count = sum(1 for r in results if r['status'] == 'NO_CONTEXT')

        avg_confidence = np.mean([r['confidence'] for r in results if r['confidence'] > 0])

        return {
            'results': results,
            'summary': {
                'total_relationships': total,
                'valid': valid_count,
                'invalid': invalid_count,
                'uncertain': uncertain_count,
                'no_context': no_context_count,
                'avg_confidence': avg_confidence,
                'method': 'similarity_based'
            }
        }

def test_similarity_validator():
    """Test the similarity-based validator"""

    print("SIMILARITY-BASED GENEALOGY VALIDATOR TEST")
    print("=" * 80)

    # Initialize validator
    validator = SimilarityBasedValidator(use_pretrained_classifier=False)

    # Test cases with known relationships
    test_relationships = [
        ("Albert Einstein", "child of", "Hermann Einstein"),
        ("Albert Einstein", "spouse of", "Mileva Marić"),
        ("Charles III", "child of", "Queen Elizabeth II"),
        ("Barack Obama", "spouse of", "Michelle Obama"),
        ("Albert Einstein", "child of", "Isaac Newton"),  # Should be invalid
        ("Barack Obama", "spouse of", "Hillary Clinton"),  # Should be invalid
        ("Hans Albert Einstein", "child of", "Albert Einstein"),
        ("Prince William", "child of", "Charles III"),
    ]

    # Test with automatic Wikipedia fetching
    results = validator.validate_multiple_relationships(
        test_relationships,
        confidence_threshold=0.25  # Lower threshold for testing
    )

    print("\n" + "=" * 80)
    print("SIMILARITY-BASED VALIDATION RESULTS")
    print("=" * 80)

    for result in results['results']:
        print(f"\nRelationship: {result['relationship']}")
        print(f"Status: {result['status']}")
        print(f"Overall Confidence: {result['confidence']:.3f}")
        print(f"Similarity Score: {result.get('similarity_score', 0):.3f}")
        print(f"Keyword Score: {result.get('keyword_score', 0):.3f}")
        print(f"Method: {result['method']}")

        if result['evidence']:
            print(f"Evidence: {result['evidence'][0][:150]}...")

        print("-" * 40)

    print(f"\nSummary Statistics:")
    summary = results['summary']
    for key, value in summary.items():
        if isinstance(value, float):
            print(f"  {key}: {value:.3f}")
        else:
            print(f"  {key}: {value}")

def test_with_same_article():
    """Test relationships within the same Wikipedia article"""

    print("\nTESTING WITH SAME ARTICLE APPROACH")
    print("=" * 50)

    validator = SimilarityBasedValidator()

    # Get Einstein's Wikipedia article
    einstein_article = validator.wiki_retriever.get_wikipedia_text("Albert Einstein")

    # Test multiple Einstein family relationships using the same article
    einstein_relationships = [
        ("Albert Einstein", "child of", "Hermann Einstein"),
        ("Albert Einstein", "child of", "Pauline Koch"),
        ("Albert Einstein", "spouse of", "Mileva Marić"),
        ("Albert Einstein", "spouse of", "Elsa Löwenthal"),
        ("Hans Albert Einstein", "child of", "Albert Einstein"),
        ("Eduard Einstein", "child of", "Albert Einstein"),
        ("Albert Einstein", "child of", "Isaac Newton"),  # Should be invalid
    ]

    print("Testing Einstein family relationships with single Wikipedia article...")

    for person1, relation, person2 in einstein_relationships:
        result = validator.validate_relationship_with_article(
            person1, relation, person2,
            article_text=einstein_article,
            confidence_threshold=0.25
        )

        print(f"{result['relationship']}: {result['status']} (confidence: {result['confidence']:.3f})")

if __name__ == "__main__":
    # Run the similarity-based test
    test_similarity_validator()

    # Test with same article approach
    test_with_same_article()
```
